````python
# Define path
csv_path = Path.cwd() / 'data' / 'weather_data_1870-1875.csv'

# Load the data
df_weather = pd.read_csv(csv_path)


# Defining a function to process the weather data
def process_weather(df):
    """Function to structure and clean weather data.
    
    Structuring includes renaming the 
    columns and selecting only observations of maximum temperature. 
    Cleaning includes inserting missing decimal, sorting and
    resetting the index and adding datetime process. 
    
    Args:
        (pd.DataFrame): dataframe with weather data. 
        
    Returns:
        (pd.DataFrame): processed weather data. 
    """

    # Structure and clean data using methods chaining.
    # Note that the original columns now are strings when loading the csv file
    # and not integers as when downloading the data.
    df_out = \
        df\
            .rename(columns={'0': 'station', '1': 'datetime', '2': 'obs_type', '3': 'obs_value'})\
            .query("obs_type == 'TMAX'")\
            .assign(obs_value=lambda df: df['obs_value']/10)\
            .sort_values(by=['station', 'datetime'], ascending=[True,False])\
            .reset_index(drop=True)\
            .copy() 

    # Area process
    df_out['area'] = df_out['station'].str[0:2]

    # Datetime process
    df_out['datetime_dt'] = pd.to_datetime(df_out['datetime'], format = '%Y%m%d')
    df_out['month'] = df_out['datetime_dt'].dt.month
    df_out['year'] = df_out['datetime_dt'].dt.year

    return df_out

# Storing the processed weather data as df_weather_period
df_weather_period = process_weather(df_weather)

# Inspect dataframe
df_weather_period.head(10)
````

````python
# Define get_colspecs function 
def get_colspecs(lines):
    """Extracts colspecs from `ghcnd-stations-column-metadata.txt`.
    
    Args:
        lines (list[str]): 
            list of relevant rows from `ghcnd-stations-column-metadata.txt` 
    
    Returns:
        (list[str]): 
            list of extracted colspecs i.e. ['1-11', '13-20', ..., '81-85']
        It strips leading whitespace of each element
    """
    colspec_idx_start = 13
    colspec_idx_end = 17 + 1  # Including idx 17
    colspecs = []
    for line in lines:
        colspec = (line[colspec_idx_start:colspec_idx_end]).strip()
        colspecs.append(colspec)
    
    return colspecs


# Apply and store get_colspecs
colspecs = get_colspecs(lines)

# Inspect colspecs
colspecs
````

> **Ex. 3.3.3.4:** Write a function named `get_colspec_pair` which takes as input a string variable named `colspec` and returns a tuple of integers. Specifically, the function should take a string similar to each element of `colspecs`, split this string by `-` and return a tuple of integers where
1. The first integer should have `1` subtracted from it (Python is 0-indexed!)
2. The second integer should stay as it is (the intervals provided to the pandas function `read_fwf` should be non-inclusive)
> As an example, applying the function to `"1-11"` and `"13-20"` should yield the following results:

```python
print(get_colspec_pair("1-11"))
## output: (0, 11)

print(get_colspec_pair("13-20"))
## output: (12, 20)
```


````python
# Define get_colspec_pair function
def get_colspec_pair(colspec):
    """ takes a string variable and returns a tuple of integers 
    by splitting at '-' and subtracting 1 from the first int.
    
    Args:
        colspec [str]: 
            the function only takes single strings
    
    Returns:
        (first int-1, second int): 
            extracted integers as a tuple i.e. (12, 20)
    """
    first, second = map(int, colspec.split('-'))
    return (first - 1, second)

# Test the function 
get_colspec_pair('1-11')
        
````

> **Ex. 3.3.3.6:** Because the fixed width column information spans the interval from `13` up and including index `17`, we know that the entries from `0` to `13` (non-inclusive) are the column names and the entries from `18` to the end of each line are the data types. Write two functions named `get_column_names` and `get_column_dtypes` which return a list of column names and a list of the data types of the columns, respectively. Likewise, apply the function `get_column_dtypes` to the `lines` variable and store the output in a variable named `column_dtypes`.

````python
# Define new function to get column names
def get_column_names(lines):
    """Extracts column names from `ghcnd-stations-column-metadata.txt`.
    
    Args:
        lines (list[str]): 
            list of relevant rows from `ghcnd-stations-column-metadata.txt` 
    
    Returns:
        (list[str]): 
            list of extracted colspecs i.e. ['ID', 'LATITUDE'...]
            It strips leading whitespace of each element
    """
    colspec_idx_start = 0
    colspec_idx_end = 12 + 1  # Including idx 17
    colspecs = []
    for line in lines:
        colspec = (line[colspec_idx_start:colspec_idx_end]).strip()
        colspecs.append(colspec)
    
    return colspecs

# Define new function to get column data types
def get_column_dtype(lines):
    """Extracts column data types from `ghcnd-stations-column-metadata.txt`.
    
    Args:
        lines (list[str]): 
            list of relevant rows from `ghcnd-stations-column-metadata.txt` 
    
    Returns:
        (list[str]): 
            list of extracted colspecs i.e. ['Character', 'Real'...]
            It strips leading whitespace of each element
    """
    colspec_idx_start = 18
    colspecs = []
    for line in lines:
        colspec = (line[colspec_idx_start:]).strip()
        colspecs.append(colspec)
    
    return colspecs

# Apply and store the names
column_names = get_column_names(lines)

# Apply and store the types
column_dtypes = get_column_dtype(lines)

# Inspect the new variables
print(column_names)
print(column_dtypes)
````

````python
import folium


def get_marker(lat, lon, station_name, avg_obs_value, icon='cloud', color="blue"):
    """Creates a `folumn.Marker` for a given station
    
    Args:
        (lat): lattitude of station
        (lon): longitude of station
        (station_name): name of station
        (avg_obs_value): avg. obs_value for given station
        
    Returns:
        (folium.Marker): object to be added to a folium map
    """
    popup = "\n".join([station_name, f"Avg. obs_value: {avg_obs_value:.2f}"])
    marker = folium.Marker(
        location=[lat, lon],
        popup=popup,
        icon=folium.Icon(icon=icon, color=color, )
    )
    return marker


# Create folium map centered on Ontario
# COORDS_ONTARIO = (51.730703, -86.938937)
COORDS_ONTARIO = (43.40168574192175, -80.33021323830818)
m = folium.Map(location=COORDS_ONTARIO, zoom_start=6)

# Zipper object to iterate through
zipper = zip(
    locations_ontario.LATITUDE,
    locations_ontario.LONGITUDE,
    locations_ontario.NAME,
    locations_ontario.station   
)


# Iterate through the zipper using the get_marker function
markers_ontario = []

for lat, lon, station_name, station_id in zipper:
    station_markers = get_marker(lat, lon, station_name, avg_obs_value_ontario[station_id])
    markers_ontario.append(station_markers)

# Add weather station markers to map 
for station_marker in markers_ontario:  
    station_marker.add_to(m)
m  # Display map
````
